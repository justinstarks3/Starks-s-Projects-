## Building a Recommendation System with Sci Kit Learns TfIdVectorizer and Cosine Similarity 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range = (1,2))
tfidf = vectorizer.fit_transform(movies['title'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

titles_list = movies["title"].tolist()


vectorizer = TfidfVectorizer(stop_words='english')


tfidf = vectorizer.fit_transform(titles_list)
def search(title):
    query_vec = vectorizer.transform([title])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity , -6)[-6:]
    indices = indices[1:]
    results = movies.iloc[indices].copy()
    results["similarity"] = similarity[indices]
    results = results.sort_values(by="similarity", ascending=False).drop(columns=["similarity"])
    return results
 
  
```

## Creating a Search Widget for Movies 

```python
import ipywidgets as widgets
from IPython.display import display

movie_input = widgets.Text(
    value='Grown Ups',
    description='Movie Title:',
    disabled=False
)
movie_list = widgets.Output()

def on_type(data):
    with movie_list:
        movie_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            display(search(title))

movie_input.observe(on_type, names='value')

display(movie_input, movie_list)
```

## Creating a Search Widget for TV Series 

```python
titles_list = shows["title"].tolist()


vectorizer = TfidfVectorizer(stop_words='english')


tfidf = vectorizer.fit_transform(titles_list)
def search(title2):
    query_vec = vectorizer.transform([title2])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity , -5)[-5:]
    results = shows.iloc[indices][::-1]
    return results 
```

```python


shows_input = widgets.Text(
    value = "The Great British Baking Show",
    description="TV Shows:",
    disabled=False
)

show_list = widgets.Output()

def on_type(data2):
    with show_list:
        show_list.clear_output()
        title2 = data2["new"]
        if len(title2) > 5:
            display(search(title2))
            
shows_input.observe(on_type , names = 'value')
display(shows_input, show_list)
```
