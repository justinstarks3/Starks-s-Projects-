```python
import pandas as pd 
df=pd.read_csv("nba_playerstats_2013.csv")
df
```

```python
df_clean= df.drop(["Age" , "Tm" , "GS" , "MP", "FG%", "3P", "3PA", "3P%", "2P" , "2PA", "2P%", "eFG%", "FT%",	"ORB","DRB", "PF", "Missed FG", "Player Efficiency" ], axis=1)
df_clean
```

```python
df["Player Efficiency Rating"]= (( df["PTS"] + df["TRB"]+ df["AST"] + df["STL"] + df["BLK"] ) - (df["TOV"]+((df["FGA"]-df["FG"]))+(df["FTA"]-df["FT"])))/df["G"]
```

```python
 df_top10_EFF=df_clean.nlargest(10, 'Player Efficiency Rating')
df_top10_EFF
```
